```python
from neuron.filters.convolve import convolve
from neuron.skullstripping import skullstripped
```

```python
import nibabel as nib
from matplotlib import pyplot as plt
```

```python
img = nib.load('data/1/T1.nii.gz')
```

```python
img_data = img.get_fdata()
```

```python
conv_img = convolve(img_data, 3)
```

```python
plt.subplot(121), plt.imshow(img_data[:,90,:], cmap=plt.cm.gray)
plt.title('Original Image')
plt.subplot(122), plt.imshow(conv_img[:,90,:], cmap=plt.cm.gray)
plt.title('Convolve Image')
```

```python
###Skullstripping example using robex
stripped, mask = skullstripped(img)
```

```python
img_data = img.get_fdata()
stripped_data = stripped.to_nibabel().get_fdata()
mask_data = mask.to_nibabel().get_fdata()
```

```python
k=90
plt.subplot(131), plt.imshow(img_data[:,k,:], cmap=plt.cm.gray), plt.title('Original')
plt.subplot(132), plt.imshow(stripped_data[:,k,:], cmap=plt.cm.gray), plt.title('Stripped')
plt.subplot(133), plt.imshow(mask_data[:,k,:], cmap=plt.cm.gray), plt.title('Mask')
```
