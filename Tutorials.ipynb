```python
from neuron.filters.convolve import convolve
```

```python
import nibabel as nib
from matplotlib import pyplot as plt
```

```python
img = nib.load('data/1/T1.nii.gz')
```

```python
img_data = img.get_fdata()
```

```python
conv_img = convolve(img_data, 3)
```

```python
plt.subplot(121), plt.imshow(img_data[:,90,:], cmap=plt.cm.gray)
plt.title('Original Image')
plt.subplot(122), plt.imshow(conv_img[:,90,:], cmap=plt.cm.gray)
plt.title('Convolve Image')
```
